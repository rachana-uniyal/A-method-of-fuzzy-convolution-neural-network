```python
%tensorflow_version 1.x

import tensorflow as tf 
from tensorflow import keras

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
BATCH_SIZE = 32
EPOCHS = 40
NUM_CLASSES = 10
LEARNING_RATE = 1e-4
MOMENTUM = 0.9
```

```python
from matplotlib import pyplot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# plot raw pixel data
	pyplot.imshow(X_test[i], cmap=pyplot.get_cmap('gray'))
# show the figure
pyplot.show()
```

```python
import cv2

for i in range (0,10000):
  X_test[i] = cv2.blur(X_test[i],(4,4))  

  
```

```python
from matplotlib import pyplot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# plot raw pixel data
	pyplot.imshow(X_test[i], cmap=pyplot.get_cmap('gray'))
# show the figure
pyplot.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, 
                                                  y_train, 
                                                  test_size=0.15, 
                                                  stratify=np.array(y_train), 
                                                  random_state=42)
```

```python
from keras.utils import np_utils

Y_train = np_utils.to_categorical(y_train, NUM_CLASSES)
Y_val = np_utils.to_categorical(y_val, NUM_CLASSES)
Y_test = np_utils.to_categorical(y_test, NUM_CLASSES)
```

```python
import numpy as np

X_train1 = np.ndarray(shape = (51000, 28,28,3), dtype = 'float32')
X_val1 =   np.ndarray(shape = (9000, 28,28,3), dtype = 'float32')
X_test1 = np.ndarray(shape = (10000,28,28,3), dtype = 'float32')
```

```python
  import cv2


  for i in range(0,51000):
      X_train1[i] = cv2.cvtColor(X_train[i],cv2.COLOR_GRAY2BGR)
      
  for i in range(0,9000):
      X_val1[i] = cv2.cvtColor(X_val[i],cv2.COLOR_GRAY2BGR)    
      
      
  for i in range(0,10000):
      X_test1[i] = cv2.cvtColor(X_test[i],cv2.COLOR_GRAY2RGB)    
```

```python
train_f = np.ndarray(shape = (51000,48,48,3), dtype = 'float32')
val_f = np.ndarray(shape = (9000,48,48,3) , dtype = 'float32')
test_f = np.ndarray(shape = (10000,48,48,3) , dtype = 'float32')
```

```python
for i in range(0,51000):
     train_f[i] = cv2.resize(X_train1[i],(48,48))
        
for i in range(0,9000):
      val_f[i] = cv2.resize(X_val1[i],(48,48))  
      
for i in range(0,10000):
     test_f[i] = cv2.resize(X_test1[i],(48,48))    
```

```python
from keras.applications.mobilenet_v2 import MobileNetV2

base_model = MobileNetV2(weights='imagenet', 
                       include_top=False, 
                       input_shape=(48, 48, 3))
```

```python
for layer in base_model.layers:
     layer.trainable = True
```

```python

from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D,BatchNormalization
from keras.engine import Model


# Extract the last layer from third block of vgg16 model
last = base_model.output


# Add classification layers on top of it
x = GlobalAveragePooling2D()(last)
x= BatchNormalization()(x)
x = Dense(256, activation='relu')(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.6)(x)
pred = Dense(NUM_CLASSES, activation='softmax')(x)
model = Model(base_model.input, pred)

```

```python
from keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.SGD(lr=LEARNING_RATE),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python

X_train = train_f
X_val =   val_f
X_test =  test_f
```

```python
print(X_test.shape)
print(X_train.shape)
print(X_val.shape)
```

```python
print(Y_test.shape)
print(Y_train.shape)
print(Y_val.shape)
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    horizontal_flip=False)
```

```python
train_datagen.fit(X_train)
train_generator = train_datagen.flow(X_train,
                                     Y_train, 
                                     batch_size=BATCH_SIZE)
```

```python
val_datagen = ImageDataGenerator(rescale=1. / 255,
    horizontal_flip=False)

val_datagen.fit(X_val)
val_generator = val_datagen.flow(X_val,
                                 Y_val,
                                 batch_size=BATCH_SIZE)
```

```python
train_steps_per_epoch = X_train.shape[0] // BATCH_SIZE
val_steps_per_epoch = X_val.shape[0] // BATCH_SIZE

history = model.fit_generator(train_generator,
                              steps_per_epoch=train_steps_per_epoch,
                              validation_data=val_generator,
                              validation_steps=val_steps_per_epoch,
                              epochs=EPOCHS,
                              verbose=1)
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
